```python
import matplotlib.pyplot as plt
import nivapy3 as nivapy
import pandas as pd
import seaborn as sn

plt.style.use("ggplot")
```

# Charts for the OSPAR Word report

Starting in 2019, NIVA is responsible for producing the OSPAR Word report. This notebook generates plots similar to those added previously from Excel.

```python
year = 2019
osp_path = f"../../../Results/OSPAR_Template/Norway 1990_{year}.xlsx"
```

## 1. Riverine inputs

```python
df = pd.read_excel(osp_path, sheet_name="RIVERINE", skiprows=5)
cols = [
    "Year",
    "RunOff",
    "Cadmium",
    "Lead",
    "Mercury",
    "Copper",
    "Zinc",
    "Ammonium",
    "Nitrate",
    "PO4-P",
    "N-Total",
    "P-Total",
    "SPM",
    "Arsenic",
    "Total Cr",
    "Nickel",
    "TOC",
]
df = df[cols]
units = list(df.iloc[0])
units_dict = {col: units[idx] for idx, col in enumerate(df.columns)}
df = df.iloc[1:]
df = df.melt(id_vars=["Year"])
df["Year"] = df["Year"].astype(int)
df["value"] = df["value"].astype(float)
df.head()
```

### 1.1. Nutrients

```python
pars = ["RunOff", "TOC", "Ammonium", "Nitrate", "PO4-P", "N-Total", "P-Total", "SPM"]
df2 = df.query("variable in @pars")

g = sn.relplot(
    data=df2,
    x="Year",
    y="value",
    col="variable",
    kind="line",
    col_wrap=2,
    marker="o",
    col_order=pars,
    aspect=1.5,
    height=4,
    facet_kws={"sharey": False, "sharex": False},
)

g.set(ylim=(0, None))

axes = g.axes.flatten()
for idx, par in enumerate(pars):
    axes[idx].set_title(par)
    axes[idx].set_ylabel(f"{par} {units_dict[par]}")
    axes[idx].set_xlabel("")

plot_path = f"../../../Results/OSPAR_Template/word_reports/{year + 1}Analysis_{year}Data/riverine_nutrients_{year}.png"
plt.savefig(plot_path, dpi=300, bbox_inches="tight")
```

### 1.2. Metals

```python
pars = [
    "Lead",
    "Cadmium",
    "Mercury",
    "Copper",
    "Zinc",
    "Arsenic",
    "Total Cr",
    "Nickel",
]
df2 = df.query("variable in @pars")

g = sn.relplot(
    data=df2,
    x="Year",
    y="value",
    col="variable",
    kind="line",
    col_wrap=2,
    marker="o",
    col_order=pars,
    aspect=1.5,
    height=4,
    facet_kws={"sharey": False, "sharex": False},
)

g.set(ylim=(0, None))

axes = g.axes.flatten()
for idx, par in enumerate(pars):
    axes[idx].set_title(par)
    axes[idx].set_ylabel(f"{par} {units_dict[par]}")
    axes[idx].set_xlabel("")

plot_path = f"../../../Results/OSPAR_Template/word_reports/{year + 1}Analysis_{year}Data/riverine_metals_{year}.png"
plt.savefig(plot_path, dpi=300, bbox_inches="tight")
```
